```python
plt.plot([2018, 2019, 2020, 2021],
         [100000, 102000, 102700, 104200])
plt.xlabel('Year')
plt.ylabel('Band revenue');
```

```python
plt.plot([2018, 2019, 2020, 2021],
         [100000, 102000, 102700, 104200])
plt.xlabel('Year')
plt.ylabel('Band revenue')
plt.ylim((0, 110000))
plt.show()
```

```python
relevant_years = []
avg_song_energy = [] 

for year in range(tracks['release_year'].min(),tracks['release_year'].max()+1):
  energy_of_all_songs_in_year = tracks.loc[tracks['release_year'] == year,'energy']
  mean_energy_for_year = energy_of_all_songs_in_year.mean()
  relevant_years.append(year)
  avg_song_energy.append(mean_energy_for_year)
    
```

```python
plt.figure(figsize=(10, 5))
plt.plot(relevant_years, avg_song_energy)
plt.title('Average energy by year')
plt.xlabel('Year')
plt.ylabel('Average energy');
```

```python
plt.plot([1, 2, 3, 4, 10], [10, 6, 12, 3, 15], 
         color='red', 
         linewidth=4, 
         linestyle='--')
plt.show()
```

```python
plt.plot([2018, 2019, 2020, 2021],
         [100000, 102000, 102700, 104200])

plt.plot([2015, 2016, 2017, 2018, 2019, 2020, 2021],
        [50000, 55000, 57000, 60000, 90000, 105000, 110000])

plt.xlabel('Year')
plt.ylabel('Band revenue')
plt.ylim((0, 120000))
plt.legend(['Flying Mice', 'Out of Band'])
plt.show()
```
