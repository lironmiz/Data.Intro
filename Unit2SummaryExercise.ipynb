3: They are about to produce an advertisement aimed at encouraging smoking cessation. The advertisement is addressed to those who define themselves as current or past smokers. Now it is necessary to decide whether the advertisement will be accompanied by a pop or rock style jingle.
For the respondents in the smoking category who answered current smoker or former smoker, which of the styles (Pop or Rock) is more popular? To decide which style is more popular, compare the number of respondents who rated it at level 4 or higher.
Summary of the guidelines:

Select the rows where the value of Smoking is current smoker or former smoker
Within the rows you selected in section 1, check how many rows have a Pop column value of 4 or higher
Within the rows you selected in section 1, check how many rows have the Rock column value of 4 or higher
The style that will be declared more popular is the one for which the number of lines received is greater

```python
smokers = young_people_survey.loc[(young_people_survey['Smoking']=='current smoker') | (young_people_survey['Smoking']=='former smoker'),:]
```

```python
print("Pop:", len(smokers.loc[smokers['Pop']>=4]))
print("Rock:", len(smokers.loc[smokers['Rock']>=4]))
```
